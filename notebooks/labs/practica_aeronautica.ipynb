### problema 2: Clasificación de Modos de Falla por Pandeo Estructural (Buckling) bajo Carga Axial

**Contexto:** Los paneles de aleación ligera y fibra de carbono que conforman el fuselaje de las etapas de un cohete sufren cargas de compresión extremas durante el despegue. Si la carga excede un límite crítico, el panel sufre pandeo elástico (Buckling). Evaluando simultáneamente el **Esfuerzo de Compresión (MPa)** y el **Desplazamiento Lateral Máximo (mm)** medido por sensores de deformación, la ingeniería estructural puede automatizar la detección del estado estructural.

**Tu Objetivo:** Desarrollar un clasificador multiclase que procese esta firma cinemática y clasifique mecánicamente el estado del panel en tres categorías: **Comportamiento Seguro/Elástico (Clase 0)**, **Pandeo Local / Inestabilidad Inicial (Clase 1)** o **Colapso Estructural Plástico / Ruptura (Clase 2)**.

#### 📋 Guía de Implementación Paso a Paso:
1. **Fase 0 (Datos):** Ejecuta la celda del instructor. El dataset asocia las mediciones mecánicas bidimensionales de esfuerzo y desplazamiento con sus respectivas etiquetas vectorizadas en One-Hot encoding.

3. **Etapa B (Compilar):** Compila el modelo utilizando el optimizador `Adam` (learning rate de `0.005`). Debido al formato One-Hot multiclase, la función de pérdida obligatoria debe ser **`'categorical_crossentropy'`**.
4. **Etapa C (Entrenar):** Lanza el entrenamiento de la red estructural por **150 épocas**.

```python
# =====================================================================
# FASE 0: SIMULACIÓN DE ENSAYOS ESTRUCTURALES DE DEFORMACIÓN POR BANDEO
# =====================================================================
np.random.seed(66)
N_ensayos = 100

# Clase 0: Seguro/Elástico (Esfuerzo moderado ~120MPa, Desplazamiento mínimo ~0.5mm)
safe_stress = np.random.normal(120.0, 10.0, N_ensayos)
safe_disp = np.random.normal(0.5, 0.1, N_ensayos)

# Clase 1: Inestabilidad Local (Esfuerzo alto estancado ~280MPa, Desplazamiento se dispara por flexión)
buckle_stress = np.random.normal(280.0, 15.0, N_ensayos)
buckle_disp = np.random.normal(4.5, 0.6, N_ensayos)

# Clase 2: Colapso Plástico (Falla del material, Esfuerzo cae bruscamente, Desplazamiento crítico)
fail_stress = np.random.normal(80.0, 12.0, N_ensayos)
fail_disp = np.random.normal(9.0, 1.0, N_ensayos)

# Consolidación de matrices mecánicas
X_structural = np.vstack([
    np.column_stack((safe_stress, safe_disp)),
    np.column_stack((buckle_stress, buckle_disp)),
    np.column_stack((fail_stress, fail_disp))
])

labels_struct = np.array([0]*N_ensayos + [1]*N_ensayos + [2]*N_ensayos)
Y_struct_onehot = keras.utils.to_categorical(labels_struct, num_classes=3)

# Mezclado aleatorio indexado
shuffle_idx = np.arange(X_structural.shape[0])
np.random.shuffle(shuffle_idx)
X_structural, Y_struct_onehot = X_structural[shuffle_idx], Y_struct_onehot[shuffle_idx]

# Partición Train/Val (80/20)
X_train_p2, X_val_p2 = X_structural[:240], X_structural[240:]
Y_train_p2, Y_val_p2 = Y_struct_onehot[:240], Y_struct_onehot[240:]

print("--> [Fase 0] Éxito: Base de datos de ensayos de fatiga mecánica cargada [240 Train, 60 Val].")
```
