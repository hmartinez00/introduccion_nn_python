### problema 2: Clasificación de Anomalías en el Subsistema de Potencia (EPS)

**Contexto:** El diagnóstico automático de fallas en órbita previene fallas catastróficas. Evaluando de forma conjunta la **Corriente del Regulador de Carga (A)** y el **Voltaje del Bus Principal (V)**, la sala de control puede determinar instantáneamente el estado de salud del satélite entre tres categorías: **Operación Nominal (Clase 0)**, **Anomalía por Cortocircuito en Paneles (Clase 1)** o **Falla por Degradación Crítica de Celdas (Clase 2)**.

**Tu Objetivo:** Desarrollar un clasificador multiclase que analice este vector bidimensional de telemetría eléctrica para alertar tempranamente fallas en el EPS.

3. **Etapa B (Compilar):** Compila la red con el optimizador `Adam` a una tasa de aprendizaje de `0.005`. Al estar las etiquetas en formato One-Hot, la función de pérdida obligatoria debe ser **`'categorical_crossentropy'`**.
4. **Etapa C (Entrenar):** Lanza el entrenamiento de tu diagnosticador por **150 épocas**.

```python
# =====================================================================
# FASE 0: SIMULACIÓN DE HISTORIAL DE FALLAS ELÉCTRICAS EN EL EPS
# =====================================================================
np.random.seed(80)
N_eventos = 100

# Clase 0: Operación Nominal (Voltaje estable ~28V, Corriente controlada ~5A)
nom_v = np.random.normal(28.0, 0.4, N_eventos)
nom_i = np.random.normal(5.0, 0.3, N_eventos)

# Clase 1: Cortocircuito en Paneles (Voltaje cae drásticamente, Corriente se dispara)
cc_v = np.random.normal(12.0, 1.2, N_eventos)
cc_i = np.random.normal(14.0, 0.8, N_eventos)

# Clase 2: Degradación Crítica de Celdas (Voltaje alto flotante por protección, Corriente ~0A)
deg_v = np.random.normal(31.5, 0.5, N_eventos)
deg_i = np.random.normal(0.2, 0.05, N_eventos)

# Consolidación de firmas eléctricas
X_eps = np.vstack([
    np.column_stack((nom_v, nom_i)),
    np.column_stack((cc_v, cc_i)),
    np.column_stack((deg_v, deg_i))
])

labels_eps = np.array([0]*N_eventos + [1]*N_eventos + [2]*N_eventos)
Y_eps_onehot = keras.utils.to_categorical(labels_eps, num_classes=3)

# Mezclado aleatorio completo
shuffler = np.arange(X_eps.shape[0])
np.random.shuffle(shuffler)
X_eps, Y_eps_onehot = X_eps[shuffler], Y_eps_onehot[shuffler]

# Partición Train/Val (80/20)
X_train_p2, X_val_p2 = X_eps[:240], X_eps[240:]
Y_train_p2, Y_val_p2 = Y_eps_onehot[:240], Y_eps_onehot[240:]

print("--> [Fase 0] Éxito: Base de datos de telemetría eléctrica vectorizada y mezclada [240 Train, 60 Val].")
```
