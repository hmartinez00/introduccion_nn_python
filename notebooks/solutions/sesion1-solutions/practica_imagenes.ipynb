### problema 2: Clasificación Espectral Multiclase de Coberturas Terrestres

**Contexto:** Los pixeles de una imagen satelital multiespectral contienen firmas de reflectancia específicas. Evaluando la respuesta en dos canales clave (Banda Roja y Banda de Infrarrojo Cercano - NIR), es posible clasificar automáticamente el tipo de cobertura terrestre en la superficie: **Agua (Clase 0)**, **Suelo Desnudo (Clase 1)** o **Vegetación Densa (Clase 2)**.

**Tu Objetivo:** Diseñar un clasificador multiclase utilizando una capa de salida con activación `softmax` y la función de pérdida apropiada para categorizar los píxeles satelitales según su comportamiento lumínico.

#### 📋 Guía de Implementación Paso a Paso:
1. **Fase 0 (Datos):** Ejecuta la celda del instructor. El dataset contiene dos variables de entrada (Roja y NIR) organizadas en una matriz de entrada, y etiquetas codificadas en One-Hot (tres columnas de salida).
2. **Etapa A (Definir):** Construye la arquitectura. La entrada recibe 2 características (`input_shape=[2]`). Agrega una capa oculta de **12 neuronas** con activación `relu`. La capa de salida debe tener **3 neuronas** (una por clase) combinadas con la función de activación **`'softmax'`** para generar la distribución de probabilidades.
3. **Etapa B (Compilar):** Configura el optimizador `Adam` (learning rate `0.005`). Debido a que las etiquetas ya están vectorizadas en formato One-Hot por el instructor, debes utilizar obligatoriamente la función de pérdida **`'categorical_crossentropy'`**.
4. **Etapa C (Entrenar):** Entrena el modelo por **120 épocas**.

```python
# =====================================================================
# FASE 0: SIMULACIÓN DE FIRMAS ESPECTRALES (ROJA Y NIR) MULTICLASE
# =====================================================================
np.random.seed(15)
N_muestras = 100

# Clase 0: Agua (Alta absorción en NIR, baja/moderada en Roja)
agua_roja = np.random.normal(0.1, 0.03, N_muestras)
agua_nir = np.random.normal(0.05, 0.02, N_muestras)

# Clase 1: Suelo Desnudo (Reflectancia moderada constante en ambas bandas)
suelo_roja = np.random.normal(0.4, 0.05, N_muestras)
suelo_nir = np.random.normal(0.45, 0.05, N_muestras)

# Clase 2: Vegetación (Baja reflectancia en Roja por clorofila, altísima en NIR)
veg_roja = np.random.normal(0.08, 0.02, N_muestras)
veg_nir = np.random.normal(0.8, 0.06, N_muestras)

# Consolidación de matrices
X_espectral = np.vstack([
    np.column_stack((agua_roja, agua_nir)),
    np.column_stack((suelo_roja, suelo_nir)),
    np.column_stack((veg_roja, veg_nir))
])

# Crear etiquetas numéricas y transformarlas a One-Hot encoding
labels = np.array([0]*N_muestras + [1]*N_muestras + [2]*N_muestras)
Y_onehot = keras.utils.to_categorical(labels, num_classes=3)

# Mezclar datos de forma aleatoria
indices = np.arange(X_espectral.shape[0])
np.random.shuffle(indices)
X_espectral, Y_onehot = X_espectral[indices], Y_onehot[indices]

# Partición Train/Val (80/20)
X_train_p2, X_val_p2 = X_espectral[:240], X_espectral[240:]
Y_train_p2, Y_val_p2 = Y_onehot[:240], Y_onehot[240:]

print("--> [Fase 0] Éxito: Firmas espectrales cargadas y codificadas en One-Hot [240 Train, 60 Val].")
```
