### problema 1: Modelado de la Curva de Transferencia y Saturación de un Amplificador (BJT/FET)

**Contexto:** Los amplificadores electrónicos basados en transistores exhiben una respuesta lineal solo dentro de un rango dinámico específico. Cuando el voltaje de entrada ($V_{in}$) excede ciertos límites, el componente entra en la zona de saturación (o corte), achatando la señal de salida ($V_{out}$) y generando distorsión armónica. Este fenómeno físico de ``clipping'' es intrínsecamente no lineal.

**Tu Objetivo:** Diseñar una Red Neuronal Profunda que aprenda a modelar la curva de transferencia de un amplificador, prediciendo con exactitud el voltaje de salida continuo e identificando de forma autónoma los puntos de saturación física.

#### 📋 Guía de Implementación Paso a Paso:
1. **Fase 0 (Datos):** Ejecuta la celda del instructor para cargar el comportamiento del circuito. Observa la meseta horizontal donde el voltaje se satura.
2. **Etapa A (Definir):** Construye la arquitectura. Para capturar el achatamiento simétrico de la curva, configura una primera capa oculta de **32 neuronas** con activación `tanh`, una segunda capa oculta de **16 neuronas** con activación `tanh`, y una neurona de salida lineal (`units=1`).
3. **Etapa B (Compilar):** Compila el modelo utilizando el optimizador `Adam` (learning rate de `0.01`) y la función de pérdida `mse`.
4. **Etapa C (Entrenar):** Entrena la red por **120 épocas** pasando las variables de entrenamiento y validación.

```python
# @title Solución Oficial - Problema 1
# =====================================================================
# FASE 0: SIMULACIÓN DE CURVA DE TRANSFERENCIA TENSIÓN-TENSIÓN
# =====================================================================
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

np.random.seed(22)

# Voltaje de entrada en milivoltios (-4.0V a 4.0V)
X_vin = np.random.uniform(-4.0, 4.0, 250)
# Física real: Amplificación lineal (Ganancia = 3) con saturación dura a +/- 9.0V
Y_vout_ideal = 3.0 * X_vin
Y_vout = np.clip(Y_vout_ideal, -9.0, 9.0) + np.random.normal(0, 0.2, 250)

# Partición de datos (80% Train, 20% Val)
X_train_p1, X_val_p1 = X_vin[:200], X_vin[200:]
Y_train_p1, Y_val_p1 = Y_vout[:200], Y_vout[200:]

print("--> [Fase 0] Éxito: Curva de transferencia del semiconductor cargada en memoria.")
```

```python
# =====================================================================
# DESAFÍO 1: TU TURNO DE ESCRIBIR CÓDIGO (MODELADO DE SATURACIÓN)
# =====================================================================

# 1. ETAPA A: Define el modelo multicapa usando activación 'tanh' para capturar la saturación simétrica
model_p1 = keras.Sequential([
    keras.layers.Dense(units=32, activation='tanh', input_shape=[1]),
    keras.layers.Dense(units=16, activation='tanh'),
    keras.layers.Dense(units=1)
])

# 2. ETAPA B: Compila el modelo (Adam 0.01, Pérdida MSE)
model_p1.compile(optimizer=keras.optimizers.Adam(learning_rate=0.01), loss='mse')

# 3. ETAPA C: Entrena el modelo por 120 épocas
history_p1 = model_p1.fit(X_train_p1, Y_train_p1, epochs=120, validation_data=(X_val_p1, Y_val_p1), verbose=0)


# Graficar la función de pérdida para el Problema 1
plt.figure(figsize=(8, 5))
plt.plot(history_p1.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history_p1.history['val_loss'], label='Pérdida de Validación')
plt.title('Función de Pérdida durante el Entrenamiento (Problema 1)')
plt.xlabel('Época')
plt.ylabel('Pérdida (MSE)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar comportamiento y auditoría
plt.figure(figsize=(6, 3.5))
plt.scatter(X_val_p1, Y_val_p1, color='#3b82f6', label='Respuesta Real Componente')
X_grid = np.linspace(-4, 4, 100)
plt.plot(X_grid, model_p1.predict(X_grid, verbose=0), color='#dc2626', lw=2.5, label='Aproximación Red (Atenuación)')
plt.title("Auditoría: Modelado de Clipping / Saturación")
plt.xlabel("Voltaje de Entrada Vin (V)")
plt.ylabel("Voltaje de Salida Vout (V)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

### problema 2: Clasificación de Integridad de Señal en Buses Digitales de Alta Velocidad

**Contexto:** En los circuitos impresos (PCB) modernos, las pistas que transportan señales digitales de alta frecuencia actúan como líneas de transmisión. Si hay desadaptación de impedancias o acoplamiento electromagnético (Crosstalk), el pulso binario se degrada. Analizando la apertura horizontal (Margen de Tiempo en ns) y la apertura vertical (Margen de Ruido en V) de un Diagrama de Ojo, podemos clasificar la salud de la línea.

**Tu Objetivo:** Desarrollar un clasificador multiclase que evalúe la integridad de la señal digital en tres estados: **Señal Limpia / Excelente (Clase 0)**, **Degradación por Jitter / Desfase (Clase 1)** o **Falla por Interferencia Interbólica / Cierre de Ojo (Clase 2)**.

#### 📋 Guía de Implementación Paso a Paso:
1. **Fase 0 (Datos):** Ejecuta la celda del instructor. El dataset combina el margen de tiempo y de ruido en un vector bidimensional de entrada, con sus respectivas etiquetas One-Hot de salida.
2. **Etapa A (Definir):** Declara la red neuronal. Agrega una capa oculta de **16 neuronas** con activación `relu`. Diseña la capa de salida con **3 neuronas** emparejadas con la función de activación **`'softmax'`** para clasificar la probabilidad de cada estado de integridad.
3. **Etapa B (Compilar):** Compila el modelo con el optimizador `Adam` (learning rate `0.005`). Configura de forma estricta la función de pérdida **`'categorical_crossentropy'`** dado el formato One-Hot de las etiquetas.
4. **Etapa C (Entrenar):** Entrena el modelo por **140 épocas**.

```python
# =====================================================================
# FASE 0: SIMULACIÓN DE MÉTRICAS DE DIAGRAMA DE OJO (INTEGRIDAD DE SEÑAL)
# =====================================================================
np.random.seed(33)
N_pistas = 100

# Clase 0: Excelente integridad (Margen de Tiempo ~0.9ns, Margen de Ruido ~3.1V)
ok_time = np.random.normal(0.9, 0.05, N_pistas)
ok_noise = np.random.normal(3.1, 0.1, N_pistas)

# Clase 1: Jitter Severo (Margen de Tiempo cae drásticamente por desfase, Ruido controlado)
jit_time = np.random.normal(0.3, 0.06, N_pistas)
jit_noise = np.random.normal(2.9, 0.15, N_pistas)

# Clase 2: Cierre de Ojo / Atrapado por Ruido (Ambos márgenes colapsan)
fail_time = np.random.normal(0.2, 0.05, N_pistas)
fail_noise = np.random.normal(0.8, 0.2, N_pistas)

# Consolidación de firmas de hardware
X_signal = np.vstack([
    np.column_stack((ok_time, ok_noise)),
    np.column_stack((jit_time, jit_noise)),
    np.column_stack((fail_time, fail_noise))
])

labels_signal = np.array([0]*N_pistas + [1]*N_pistas + [2]*N_pistas)
Y_signal_onehot = keras.utils.to_categorical(labels_signal, num_classes=3)

# Mezclado aleatorio indexado
mix_idx = np.arange(X_signal.shape[0])
np.random.shuffle(mix_idx)
X_signal, Y_signal_onehot = X_signal[mix_idx], Y_signal_onehot[mix_idx]

# Partición Train/Val (80/20)
X_train_p2, X_val_p2 = X_signal[:240], X_signal[240:]
Y_train_p2, Y_val_p2 = Y_signal_onehot[:240], Y_signal_onehot[240:]

print("--> [Fase 0] Éxito: Base de datos de métricas de alta frecuencia cargada y lista [240 Train, 60 Val].")
```

```python
# @title Solución Oficial - Problema 2
# =====================================================================
# DESAFÍO 2: TU TURNO DE ESCRIBIR CÓDIGO (AUDITOR DE INTEGRIDAD DE SEÑAL)
# =====================================================================

# 1. ETAPA A: Define el modelo clasificador (2 características de entrada, salida con 3 neuronas Softmax)
model_p2 = keras.Sequential([
    keras.layers.Dense(units=16, activation='relu', input_shape=[2]),
    keras.layers.Dense(units=3, activation='softmax')
])

# 2. ETAPA B: Compila el modelo (Adam 0.005, Categorical Crossentropy)
model_p2.compile(optimizer=keras.optimizers.Adam(learning_rate=0.005), loss='categorical_crossentropy', metrics=['accuracy'])

# 3. ETAPA C: Entrena el modelo por 140 épocas
history_p2 = model_p2.fit(X_train_p2, Y_train_p2, epochs=140, validation_data=(X_val_p2, Y_val_p2), verbose=0)

# Evaluación estadística del rendimiento
loss_f, acc_f = model_p2.evaluate(X_val_p2, Y_val_p2, verbose=0)
print(f"--> [ÉXITO] Clasificador de hardware operativo.")
print(f"    Porcentaje de precisión en diagnóstico de pistas: {acc_f*100:.2f}%")
```

```python
# Graficar la función de pérdida y la precisión para el Problema 2
plt.figure(figsize=(12, 5))

# Subplot para la pérdida
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, primer gráfico
plt.plot(history_p2.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history_p2.history['val_loss'], label='Pérdida de Validación')
plt.title('Función de Pérdida durante el Entrenamiento (Problema 2)')
plt.xlabel('Época')
plt.ylabel('Pérdida (Categorical Crossentropy)')
plt.legend()
plt.grid(True)

# Subplot para la precisión
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, segundo gráfico
plt.plot(history_p2.history['accuracy'], label='Precisión de Entrenamiento')
plt.plot(history_p2.history['val_accuracy'], label='Precisión de Validación')
plt.title('Precisión durante el Entrenamiento (Problema 2)')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()
plt.grid(True)

plt.tight_layout() # Ajustar el diseño para evitar superposiciones
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Crear una malla (mesh grid) para cubrir el espacio de características
x_min, x_max = X_signal[:, 0].min() - 0.1, X_signal[:, 0].max() + 0.1
y_min, y_max = X_signal[:, 1].min() - 0.5, X_signal[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# 2. Predecir la clase para cada punto en la malla
grid_points = np.c_[xx.ravel(), yy.ravel()]
predictions = model_p2.predict(grid_points, verbose=0)
Z = np.argmax(predictions, axis=1)
Z = Z.reshape(xx.shape)

# 3. Graficar los resultados
plt.figure(figsize=(10, 6))
# Fondo con las regiones de decisión
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

# Puntos de entrenamiento reales
scatter = plt.scatter(X_train_p2[:, 0], X_train_p2[:, 1], c=np.argmax(Y_train_p2, axis=1),
            edgecolors='k', cmap='viridis', label='Datos de Entrenamiento')

plt.title("Visualización de Fronteras de Decisión - Integridad de Señal")
plt.xlabel("Margen de Tiempo (ns)")
plt.ylabel("Margen de Ruido (V)")

# Leyenda personalizada
handles, _ = scatter.legend_elements()
plt.legend(handles, ['Clase 0: Excelente', 'Clase 1: Jitter', 'Clase 2: Falla'], title="Categorías")
plt.grid(True, alpha=0.2)
plt.show()
```
